# Exploratory Data Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('data/train.csv')
print("Initial shape of the dataset:", df.shape)
df.head()
```

```python
df = df[['id', 'VotedHelpful', 'TotalVotes', 'summary', 'reviewText', 'genres', 'Score', 'album_mbid', 'reviewerID', 'artist_mbid']]
df['genres'] = df['genres'].str.split(',')
```

```python
df.isnull().sum()
```

```python
df[df['Score'].isnull()].head()
```

```python
unique_genres = df['genres'].explode().unique()
print("Number of unique genres:", len(unique_genres))
print("Unique genres:", unique_genres)
```

```python
for score in range(1, 6):
    print(f"Examples with Score {score}:")
    print(df[df['Score'] == score].head())
```
